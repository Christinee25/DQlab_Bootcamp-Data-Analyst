# **2. Pandas Structure**

## A. Series

Library pandas memiliki dua jenis struktur data, yaitu series dan dataframe. Struktur data series merupakan satu kolom bagian dari tabel data frame yang merupakan 1 dimensional numpy array sebagai baris datanya, terdiri dari 1 tipe data.

image.png

````python
#Series
sales = [12, 34, 21, 34]

sales_series = pd.Series(sales)

display(sales_series)
````

````python
#Inisialisasi data series
cust_sales = {
     'cust1' : 12,
     'cust2' : 34,
     'cust3' : 21,
     'cust4' : 34
 }

#convert to series type
sales_series = pd.Series(cust_sales)

display(sales_series)
````

##**C. Sorting**

Sorting biasanya digunakan untuk mengurutkan row data secara ascending (dari A-Z) atau descending (dari Z-A). Code yang dapat digunakan
```
DataFrame.sort_values(by, ascending=True, ignore_index=False)
```

Keterangan:
* by : satu kolom yang ingin di urutkan
* ascending : true jika ascending, false jika descending
* ignore_index : True jika masih ingin indexnya terurut, false jika sebaliknya

````python
#sort data berdasarkan order id
display(data.sort_values('Order ID').head())

#sort data berdasarkan order id
data.sort_values(['Category', 'Sub-Category'], ignore_index=True)

#sort data sekaligus mengubah data menjadi hasil sintaks sort secara permanen
data.sort_values(['Category', 'Sub-Category'], ignore_index=True, inplace = True)
````

## F. Agregasi Data

Agregasi adalah serangkaian proses yang dilakukan untuk mengumpulkan data dan dinyatakan dalam bentuk ringkasan (misal total dikelompokkan berdasarkan kategori tertentu)
```
Data.Frame.group_by(by, as_index).agg(aggregation_column = (column, aggregation_type))
```
Agrregation Type:
* .sum
* .mean
* .max
* .min
* .nunique
* .count

````python
#Hitung total profit berdasarkan order id
data.groupby(by='Order ID').agg(total_profit=('Profit','mean')).head()
data.groupby(by='Order ID').agg(total_profit=('Profit','sum'), total_sales = ('Sales','sum')).head(10)
data_group_by = data[['Order ID', 'Profit', 'Sales']].groupby(by='Order ID').sum().head(10)
data_group_by
````

## G. Iteration

Iterasi digunakan untuk membaca satu per satu dari row data yang ada. iterasi yang dapat digunakan antara lain:
> DataFrame.columns

> DataFrame.index

> DataFrame.iterrows()


````python
print(data.columns)

#Menggunakan iterasi
for i in data.columns:
  print(i)
````

````python
print(data.index)

for i in data.index:
  print(data['Order ID'][i], data['Sales'][i])
````

````python
for i, j in data.iterrows():
  print(j['Order ID'], j['Sales'])
````
