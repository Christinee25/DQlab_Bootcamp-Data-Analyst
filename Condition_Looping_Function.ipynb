*   Jika kurang dari 5% dari total karyawan maka dikategorikan rendah.

Seorang investor memiliki modal awal sebesar Rp 100 juta dan ingin menginvestasikan uangnya dalam bentuk deposito dengan bunga tetap 5% per tahun. Dia ingin mengetahui berapa total nilai investasinya setelah beberpaa tahun.

Buatlah fungsi nilai_investasi yang menerima parameter modal_awal, bunga_tahunan, dan tahun, dan mengembalikan nilai investasi setelah sejumlah tahun tertentu.

```python
modal_awal = 100

def nilai_investasi(modal_awal, tahun, bunga_tahunan=0.5):
  total_investasi = modal_awal + (modal_awal * bunga_tahunan * tahun)
  print('nilai investasi', total_investasi)
# Ex usage:
nilai_investasi(modal_awal, tahun=5)
```

```python
def nilai_investasi(modal_awal, bunga_tahunan, tahun):
  investasi = modal_awal
  for i in range(tahun):
    investasi += investasi * (bunga_tahunan/100)
  return investasi

# Input dari pengguna
modal_awal = float(input("Masukkan modal awal: Rp"))
bunga_tahunan = float(input("Masukkan bunga tahunan (%): "))
tahun = int(input("Masukkan jumlah tahun: "))

# Menghitung nilai investasi menggunakan fungsi
hasil_investasi = nilai_investasi(modal_awal, bunga_tahunan, tahun)

# Menampilkan hasil
print(f"Setelah {tahun} tahun, nilai investasi menjadi: Ro {hasil_investasi:.2f}")
```

### Challenge 3-3

--Program KPR--


Anda diminta untuk membantu membuatkan program KPR kepada nasabah. Pemberian bunga KPR diberikan berdasarkan kategori nasabah sebagai berikut:


*   Solitaire (Saldo mmengendap > 10 Juta) mendapat rate 2.1%
*   Prioritas (Saldo mengendap > 1 Juta) mendapat rate 3.0%
*   Nasabah K1 (Saldo mengendap > 500 Ribu) mendapat rate 4.2%
*   Nasabah Reguler (Saldo mengendap <= 500 Ribu) mendapat rate 4.7%

Hitung total cicilan yang harus dibayarkan selama sebulan, berdasarkan kategori tenor cicilan 10 tahun/ 15 tahun/ 20 tahun



```python
# Fungsi untuk Menghitung Cicilan
def hitung_cicilan(nasabah, harga_rumah, tenor):
  if nasabah=='solitaire':
    rate = 0.021
  elif nasabah == 'prioritas':
    rate = 0.03
  elif nasabah == 'k1':
    rate = 0.042
  else:
    rate = 0.047
  cicilan = (harga_rumah+(harga_rumah*rate))/(tenor*12)
  return cicilan, rate

# Fungsi untuk menampilkan input output
def kpr():
  ## Define variabel input
  saldo_mengendap = int(input('Masukkan total saldo mengendap: '))
  harga_rumah = float(input('Masukkan harga rumah idaman: '))
  tenor = int(input('Tenor Cicilan (10/15/20 Tahun): '))

  ## Validasi tenor pilihan
  tenor_list = [10,15,20]
  if tenor not in tenor_list:
    print('Tenor anda tidak sesuai.')
    return

  if saldo_mengendap <= 500000:
    nasabah = 'reguler'
    cicilan, rate = hitung_cicilan(nasabah, harga_rumah, tenor)
  elif saldo_mengendap <= 1000000:
    nasabah = 'k1'
    cicilan, rate = hitung_cicilan(nasabah, harga_rumah, tenor)
  elif saldo_mengendap <= 10000000:
    nasabah = 'prioritas'
    cicilan, rate = hitung_cicilan(nasabah, harga_rumah, tenor)
  else:
    nasabah = 'solitaire'
    cicilan, rate = hitung_cicilan(nasabah, harga_rumah, tenor)

  cicilan = round(cicilan, 2)
  rate = round(rate*100,1)
  print(f'Anda merupakan nasabah {nasabah} dan memperoleh bunga {rate}%')
  print(f'Cicilan perbulan: Rp {cicilan}')
```
